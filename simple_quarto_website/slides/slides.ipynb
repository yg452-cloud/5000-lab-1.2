```python
# | code-fold: true

import torch
import numpy as np
import matplotlib.pyplot as plt

# Simple Vision Transformer demo
def simple_attention(image_patches):
    """Apply basic attention to image patches"""
    # Random attention weights (simplified)
    attention = torch.softmax(torch.randn(16, 16), dim=-1)
    return attention

# Create sample image data
image_size = 64
patch_size = 16
num_patches = (image_size // patch_size) ** 2

# Generate sample image patches
patches = torch.randn(num_patches, 3, patch_size, patch_size)

# Apply attention
attention_weights = simple_attention(patches)

# Visualize results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Show attention heatmap
im1 = ax1.imshow(attention_weights.detach().numpy(), cmap='viridis')
ax1.set_title('Attention Weights')
ax1.set_xlabel('Patch Index')
ax1.set_ylabel('Patch Index')
plt.colorbar(im1, ax=ax1)

# Show sample patches
patch_grid = torch.cat([patches[i] for i in range(4)], dim=2)
patch_grid = (patch_grid - patch_grid.min()) / (patch_grid.max() - patch_grid.min())
ax2.imshow(patch_grid.permute(1, 2, 0))
ax2.set_title('Sample Image Patches')
ax2.axis('off')

plt.tight_layout()
plt.show()

print(f"Number of patches: {num_patches}")
print(f"Patch size: {patch_size}x{patch_size}")
print(f"Attention matrix shape: {attention_weights.shape}")
```
